### <font color=red>Vertical Bar Chart Plotting</font>
* **Example**





```python
import matplotlib.pyplot as plot

students=["Emre","Esma","Ahmet","Demet","Kerem"]
grades=[90,30,45,100,87]

plot.bar(students,grades, align="center", color="b", alpha=0.8)

plot.ylabel("Score")
plot.title("Exam Grades")

plot.show()
```

### <font color=red>Vertical Bar Chart Plotting</font>
* **Example**

```python
import matplotlib.pyplot as plot

students=["Emre","Esma","Ahmet","Demet","Kerem"]
grades=[90,30,45,100,87]


plot.barh(students,grades,align="center",color="g",alpha=0.5)

plot.xlabel("Score")
plot.title("Exam Grades")

plot.show()
```
